```python
import os
import numpy as np
import time
import subprocess
```

```python
# manipulating earthquake data
fname = 'hypoDD/bootstrap/hypoDD.pha'
data = open(fname)
with open(fname) as f:
    content = f.read().splitlines()

firstLine = data.readlines()
for i in range(len(firstLine)):
	firstLine[i]=firstLine[i].split()
data.close()

#standard deviation of pick-time in seconds
std = 0.1

# number of realization
N = 50

for nn in range(N):
    #write DATA
    hypodd = open('hypoDD/bootstrap/hypoDD'+str(nn)+'.pha', 'w+')
    # hypodd.write(content[0]+'\n')
    j = 1
    while j < len(content)-1:
        hypodd.write(content[j - 1]+'\n')
        i = 0
        while firstLine[j][0] != '#':
            i += 1
            j += 1
            if j == len(content):
                break
        j += 1

        # Generating random number with normal distribution
        rand = np.random.normal(loc = 0, scale = std, size = i)

        k = j-i-1
        count = np.arange(k, j - 1, 1)
        r = 0
        for m in count:
            tp = np.round(float(firstLine[m][1]) - rand[r], decimals=2)
            hypodd.write('%5s'%firstLine[m][0]+'%12s'%str(tp)+'%8s'%firstLine[m][2]+'%4s'%firstLine[m][3]+'\n')
            r += 1

    hypodd.close()

    time.sleep(2)
    # modify and run ph2dt
    print('===================================')
    print('===================================')
    print('=========Now running ph2dt=========')
    print('===================================')
    print('===================================')
    with open('hypoDD/bootstrap/ph2dt.inp', 'r') as file:
        # read a list of lines into data
        data = file.readlines()

    # now change the line
    data[4] = 'hypoDD'+str(nn)+'.pha\n'

    # and write everything back
    with open('hypoDD/bootstrap/ph2dt.inp', 'w') as file:
        file.writelines( data )
        
    directory = "hypoDD/bootstrap"
    # Command untuk navigasi ke direktori dan menjalankan file
    command = f"cd {directory} && wsl ./ph2dt ph2dt.inp"
    # Running the command
    process = subprocess.Popen(command, shell=True, stdout=subprocess.PIPE, stderr=subprocess.PIPE)

    time.sleep(15)
    print('===================================')
    print('===================================')
    print('=========Now running hypoDD========')
    print('===================================')
    print('===================================')

    with open('hypoDD/bootstrap/hypoDD.inp', 'r') as file:
        # read a list of lines into data
        data = file.readlines()

    # now change the line
    data[18] = 'hasil/hypoDD'+str(nn)+'.reloc\n'

    # and write everything back
    with open('hypoDD/bootstrap/hypoDD.inp', 'w') as file:
        file.writelines( data )


    directory = "hypoDD/bootstrap"
    # Command untuk navigasi ke direktori dan menjalankan file
    command = f"cd {directory} && wsl ./hypodd hypoDD.inp"
    # Running the command
    process = subprocess.Popen(command, shell=True, stdout=subprocess.PIPE, stderr=subprocess.PIPE)
    
    time.sleep(15)
```

```python
import pandas as pd
from obspy import UTCDateTime
import matplotlib.pyplot as plt
import numpy as np
file_path = 'chile2014.txt'
# Load the data into a pandas DataFrame, skipping the header lines
df = pd.read_csv(file_path, 
                 delim_whitespace=True, 
                 skiprows=21, 
                 names=['year', 'month', 'day', 'hour', 'minute', 'second', 'longitude', 'latitude', 'depth_km', 'magnitude'])

# Convert the date and time columns into a single datetime column
df['time'] = pd.to_datetime(df[['year', 'month', 'day', 'hour', 'minute', 'second']])

# Drop the original year, month, day, hour, minute, and second columns
df = df.drop(['year', 'month', 'day', 'hour', 'minute', 'second'], axis=1)

# Rearrange the columns so that datetime is the first column
df = df[['time', 'longitude', 'latitude', 'depth_km', 'magnitude']]

df
```

```python
filtered_df = df[df['time'] < '2014-04-16']
filtered_df
```
